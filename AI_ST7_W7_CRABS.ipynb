```python
import pandas as pd
url = '/content/crabs.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['fontallip']),y=data['depth'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la: ',model.coef_)

a = model.score(X=pd.DataFrame(data['fontallip']),y=data['depth'])
print("Accuracy is",round(a*100),"%")

y_pred = model.predict(X=pd.DataFrame(data['fontallip']))

data.plot(kind='scatter',x='fontallip',y='depth',figsize=(9,9),color='green', xlim=(0,25))
plt.plot(data['fontallip'],y_pred,color='red')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['fontallip'],data['fontallip']**2]).T
model.fit(X, y=data['depth'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coefficient cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:',round((model.score(X, y=data['depth']))*100,3),'%')

import numpy as np
range = np.arange(5,25.1,0.1)
X_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind ='scatter',x='fontallip',y='depth',figsize = (9,9),color='green',xlim=(5,25))
plt.plot(range,y_pred,color='red')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['fontallip'],data['fontallip']**2,data['fontallip']**3]).T
model.fit(X, y=data['depth'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coefficient cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:',round((model.score(X, y=data['depth']))*100,3),'%')

import numpy as np
range = np.arange(5,25.1,0.1)
X_range = pd.DataFrame([range, range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind ='scatter',x='fontallip',y='depth',figsize = (9,9),color='green',xlim=(5,25))
plt.plot(range,y_pred,color='red')
```
