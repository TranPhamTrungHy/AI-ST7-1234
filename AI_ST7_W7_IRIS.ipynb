```python
import pandas as pd
url = '/content/iris.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)

```

```python
import seaborn as sns
sns.heatmap(correlation, annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['petal.length']),y=data['petal.width'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la: ',model.coef_)

a = model.score(X=pd.DataFrame(data['petal.length']),y=data['petal.width'])
print("Accuracy is",round(a*100,3),"%")

y_pred = model.predict(X=pd.DataFrame(data['petal.length']))

data.plot(kind='scatter',x='petal.length',y='petal.width',figsize=(9,9),color='green', xlim=(1,8))
plt.plot(data['petal.length'],y_pred,color='red')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
X=pd.DataFrame([data['petal.length'],data['petal.length']**2]).T
model.fit(X,y=data['petal.width'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la: ',model.coef_)

a = model.score(X,y=data['petal.width'])
print("Accuracy is",round(a*100,3),"%")


import numpy as np
range = np.arange(1,8,0.1)
X_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind ='scatter',x='petal.length',y='petal.width',figsize = (9,9),color='green',xlim=(1,8))
plt.plot(range,y_pred,color='red')
```
